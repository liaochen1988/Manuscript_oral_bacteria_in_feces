```python
import pandas as pd
import numpy as np
```

```python
df_tr=pd.read_csv("../bact_refseq_TR.csv")
df_tr.columns=['ASV_TR','Sequence']
df_tr.head()
```

```python
df_cl=pd.read_csv("tblASVtaxonomy_silva138_bact.csv")
df_cl = df_cl[['ASV','Sequence']]
df_cl.columns = ['ASV_CL','Sequence']
df_cl.head()
```

```python
df = pd.merge(df_tr, df_cl, left_on='Sequence', right_on='Sequence', how='outer')
df['ASV'] = df['ASV_CL']
df.loc[df['ASV_CL'].isnull(),'ASV'] = [x.split(';')[0]+'_TR' for x in df.loc[df['ASV_CL'].isnull(),'ASV_TR']]
assert len(df[df.ASV_CL.isnull()]) == len(set(df[df.ASV_CL.isnull()].ASV))
df
```

```python
ofile = open("bact_seqs.fasta", "w")

for name, seq in zip(df.ASV, df.Sequence):
    ofile.write(">" + name + "\n" +seq + "\n")
ofile.close()
```

```python
df_tax = pd.read_csv("tblASVtaxonomy_silva138_bact.csv")
df_tax
```

```python
df_taxa = pd.read_csv("../Silva/tax_silva.csv")
df_boot = pd.read_csv("../Silva/boot_silva.csv")
df_boot.columns= ['Confidence'+x for x in df_boot.columns]
df_new = pd.merge(df_taxa, df_boot, left_index=True, right_index=True, how='inner')
df_new = pd.merge(df[['ASV','Sequence']].set_index('ASV'), df_new, left_index=True, right_index=True, how='inner')
df_new.index.name='ASV'
df_new
```

```python
df_new.to_csv("tblASVtaxonomy_silva138_bact.csv")
```

```python
df_tax = pd.read_csv("tblASVtaxonomy_silva138_bact.csv")
df_count = pd.read_csv("tblcounts_melt_bact.csv")
assert len(set(df_count.ASV)-set(df_tax.ASV))==0

asvs_to_remove = list(set(df_tax.ASV)-set(df_count.ASV))
df_tax = df_tax[~df_tax.ASV.isin(asvs_to_remove)]
df_tax.to_csv("tblASVtaxonomy_silva138_bact.csv", index=False)
```
