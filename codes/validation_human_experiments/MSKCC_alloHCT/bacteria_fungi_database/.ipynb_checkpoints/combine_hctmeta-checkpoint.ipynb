```python
df = pd.read_csv('../tblhctmeta.csv', index_col=0)
df = df[df.notnull().any(axis=1)]
df = df[['MRN','Sex','BMT','m.Disease','m.Age','m.Source','engraftment_date','death_date']]
df = df[~df.MRN.str.contains('anon')]
df.MRN = df.MRN.astype(int)
df_key = pd.read_csv('../tCastoriPatientDictionaryWithDatesAndOtherStudiesPatientId.txt').rename({'cmicId':'PatientID'}, axis=1)
df = pd.merge(df, df_key[['MRN','PatientID','referenceDate']], left_on=['MRN'], right_on=['MRN'], how='inner')
df = df.drop('MRN', axis=1)
df.loc[df.PatientID=='952','BMT'] = '2014-12-17'
df['BMT'] = pd.to_datetime(df['BMT'], format='%Y-%m-%d')
df['death_date'] = pd.to_datetime(df['death_date'], format='%Y-%m-%d')
df['engraftment_date'] = pd.to_datetime(df['engraftment_date'], format='%Y-%m-%d')
df['referenceDate'] = pd.to_datetime(df['referenceDate'], format='%Y-%m-%d %H:%M:%S.%f')
df['TimepointOfTransplant'] = (df['BMT']-df['referenceDate']).dt.days
df['TimepointOfDeath'] = (df['death_date']-df['referenceDate']).dt.days
df['EngraftmentDayRelativeToNearestHCT'] = (df['engraftment_date']-df['BMT']).dt.days
df = df[(df.PatientID.notnull())]
df = df.drop(['engraftment_date','death_date','referenceDate','BMT'], axis=1)
df.columns=['Sex','Disease','Age','HCTSource','PatientID','TimepointOfTransplant','TimepointOfDeath','EngraftmentDayRelativeToNearestHCT']
df['Disease'] = df.Disease.replace({'Leukemia (Second Primary)':'Leukemia',
                                    'Myelodysplastic Syndromes (Second Primary)':'Myelodysplastic Syndromes',
                                    'Myelodysplastic Syndrome':'Myelodysplastic Syndromes',
                                    "Non-Hodgkin's Lymphoma (Second Primary)":"Non-Hodgkin's Lymphoma",
                                    'Myeloproliferative Disorder (Second Primary)':'Myeloproliferative Disorder'
                                   })
df['HCTSource'] = df.HCTSource.replace({'Allo BM Unmodified':'BM_unmodified',
                                        'Allo PBSC Unmodified Only':'PBSC_unmodified',
                                        'Allo Cord Blood':'cord',
                                        'Allo PBSC CD34+ (CliniMacs/Miltenyi)':'TCD',
                                        'Allo Double Cord Blood':'cord',
                                        'Allo PBSC CD34+/Double Cord Blood':'cord',
                                        'Allo BM SBA-E-':'BM_unmodified',
                                        'Allo PBSC CD34+E- Only':'TCD',
                                        'Allo PBSC CD34+ (CliniMacs/Miltenyi) and BM SBA-E-':'BM_unmodified',
                                        'Allo PBSC CD34+/SINGLE Cord Blood':'cord',
                                        'Allo PBSC alpha/beta T-Cell and PBSC CD34+':'TCD',
                                        'Allo PBSC & BM Unmodified':'BM_unmodified'
                                       })
df.head()
```
