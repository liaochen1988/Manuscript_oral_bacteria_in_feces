```python
import pandas as pd
import numpy as np
```

# Read wbc data

```python
dfb = pd.read_csv("../tblwbc_Brad.csv")
dfb['Unit'] = 'K_per_uL'
dfb = dfb[['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT']]
dfb['Priority_BT'] = 2
dfb.head()
```

```python
set(dfb.BloodCellType)
```

```python
dft = pd.read_csv("../tblwbc_tr.csv", index_col=0)
dft = dft[['MRN','test_date','WBC','Abs.Neut','Abs.Lymph','test_bmtday']]
dft.columns = ['MRN','Timepoint','WBCtotal','Neutrophils','Lymphocytes','DayRelativeToNearestHCT']
dft.MRN = dft.MRN.astype(int)
df_key = pd.read_csv('tCastoriPatientDictionaryWithDatesAndOtherStudiesPatientId.txt').rename({'cmicId':'PatientID'}, axis=1)
dft = pd.merge(dft, df_key[['MRN','PatientID','referenceDate']], left_on=['MRN'], right_on=['MRN'], how='inner')
dft = dft.drop('MRN', axis=1)
dft['Timepoint'] = pd.to_datetime(dft['Timepoint'], format='%Y-%m-%d')
dft['referenceDate'] = pd.to_datetime(dft['referenceDate'], format='%Y-%m-%d %H:%M:%S.%f')
dft['Timepoint'] = (dft['Timepoint']-dft['referenceDate']).dt.days
dft = dft.drop('referenceDate', axis=1)
dft = dft.set_index(['PatientID','Timepoint','DayRelativeToNearestHCT']).stack().reset_index()
dft.columns = ['PatientID','Timepoint','DayRelativeToNearestHCT','BloodCellType','Value']
dft['Unit'] = 'K_per_uL'
dft = dft[['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT']]
dft['Priority_TR'] = 1
dft.head()
```

```python
set(dft.BloodCellType)
```

```python
dfj = pd.read_csv("../tblbc_js.csv")
dfj.Value = dfj.Value.astype(str)
dfj['Priority_JS'] = 3
dfj.head()
```

```python
len(dfj)
```

# Join tables

```python
df = pd.merge(dfj, dfb,
              left_on=['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT'],
              right_on=['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT'],
              how='outer'
             )
df = pd.merge(df, dft,
              left_on=['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT'],
              right_on=['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT'],
              how='outer'
             )
df[['Priority_JS','Priority_BT','Priority_TR']]=df[['Priority_JS','Priority_BT','Priority_TR']].fillna(0)
df['Priority'] = df[['Priority_JS','Priority_BT','Priority_TR']].max(axis=1)
df = df.sort_values(['PatientID','Timepoint','BloodCellType','Priority'])
df = df.loc[df[['PatientID','Timepoint','BloodCellType','Priority']].drop_duplicates().index]
df = df[['PatientID','Timepoint','BloodCellType','Value','Unit','DayRelativeToNearestHCT']]
df.to_csv("tblbc.csv", index=False)
df
```
