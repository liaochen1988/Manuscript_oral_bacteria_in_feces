```python
import pandas as pd
import numpy as np
```

```python
df_old = pd.read_csv("../JY_scientific_data_paper/tblASVsamples.csv")
df_old['SampleType'] = 'Stool'
df_old.loc[df_old.SampleID=='1778A','Consistency']= 'semi-formed'
df_old.loc[df_old.SampleID=='FMT.0249D','Consistency'] = 'semi-formed'
df_old = df_old[['SampleID','PatientID','Timepoint','Consistency','SampleType']]
df_old.head()
```

```python
df = pd.read_csv("../sample_tr.csv", index_col=0)
df = df[['MRN','Sample_ID','DateCollection','SampleType','Consistency']]
df = df[(df.MRN.notnull()) & (df.MRN!=0)]
df.MRN = df.MRN.astype(str)
df = df[~df.MRN.str.contains('anon')]
df.MRN = df.MRN.astype(int)
df_key = pd.read_csv('../tCastoriPatientDictionaryWithDatesAndOtherStudiesPatientId.txt').rename({'cmicId':'PatientID'}, axis=1)
df = pd.merge(df, df_key[['MRN','PatientID','referenceDate']], left_on=['MRN'], right_on=['MRN'], how='inner')
df = df.drop('MRN', axis=1)
df['DateCollection'] = pd.to_datetime(df['DateCollection'], format='%Y-%m-%d')
df['referenceDate'] = pd.to_datetime(df['referenceDate'], format='%Y-%m-%d %H:%M:%S.%f')
df['Timepoint'] = (df['DateCollection']-df['referenceDate']).dt.days
df = df[['Sample_ID','SampleType','PatientID','Timepoint','Consistency']]
df.columns=['SampleID','SampleType','PatientID','Timepoint','Consistency']
df.Consistency = df.Consistency.replace({"formed stool":"formed","semi-formed, jelly":"semi-formed","thick liquid":"liquid",
                                         "Liquid":"liquid","stool":"formed","soft/formed (pink)":"semi-formed",
                                        "Formed stool":"formed","soft/unformed (pink)":"semi-formed","hard (pink)":"formed",
                                        "liquid (pink)":"liquid","Semi-formed":"semi-formed","for7/5":np.NaN})
df
```

```python
df_new = pd.merge(df, df_old, 
                  left_on=['SampleID','SampleType','PatientID','Timepoint','Consistency'],
                  right_on=['SampleID','SampleType','PatientID','Timepoint','Consistency'],
                  how='outer')
df_new = df_new.drop_duplicates()
for k,sid in enumerate(df_new.SampleID):
    if sid in []:
        continue
    df_ = df_new[df_new.SampleID==sid]
    if len(df_)>1:
        #break
        assert 0
```

```python
df_new.to_csv("tblASVsamples.csv", index=False)
```
