```python
import pandas as pd
import os
import numpy as np
from copy import deepcopy
```

```python
df = pd.read_csv("tblInfections_Thierry.csv", index_col=0)
df = df[['pos','result','result_full','MRN','BMT','test_bmtday']]
df_key = pd.read_csv('../tCastoriPatientDictionaryWithDatesAndOtherStudiesPatientId.txt').rename({'cmicId':'PatientID'}, axis=1)
df = pd.merge(df, df_key[['MRN','PatientID','referenceDate']], left_on=['MRN'], right_on=['MRN'], how='inner')
df = df.drop('MRN', axis=1)
df['BMT'] = pd.to_datetime(df['BMT'], format='%Y-%m-%d')
df['referenceDate'] = pd.to_datetime(df['referenceDate'], format='%Y-%m-%d %H:%M:%S.%f')
df['Timepoint'] = (df['BMT']-df['referenceDate']).dt.days+df['test_bmtday']
df = df[['PatientID','Timepoint','result','result_full','test_bmtday','pos']]
df.columns = ['PatientID','Timepoint','InfectiousAgent','Full','DayRelativeToNearestHCT','Positive']
df['InfectiousAgent'] = [x.replace('.','').replace('\r','').strip().rstrip('\n').strip().capitalize() if str(x)!='nan' else np.NaN for x in df['InfectiousAgent']]
df = df.reset_index(drop=True)
df2 = deepcopy(df)
index_to_remove = []
for k,ia in enumerate(df['InfectiousAgent']):
    if str(ia)=='nan':
        continue
    if '\n' in ia:
        tmps = ia.split('\n')
        for ia2 in tmps:
            df2.loc[len(df2)] = df.iloc[k]
            df2.loc[len(df2)-1,'InfectiousAgent'] = ia2
        index_to_remove.append(k)
df2 = df2.loc[[idx for idx in df2.index if idx not in index_to_remove]]
df2['InfectiousAgent'] = [x.replace('.','').replace('\r','').strip().rstrip('\n').strip().capitalize() if str(x)!='nan' else np.NaN for x in df2['InfectiousAgent']]
df2 = df2[df2.Positive==True]
df2.to_csv('intermediate_results.csv', index=False)
df2.head()
```

```python
df = pd.read_csv('intermediate_results.csv')
df = df[~((df.InfectiousAgent.notnull()) & (df.Positive==True) & (df.InfectiousAgent.str.contains('Culture negative')))]
df.loc[(df.InfectiousAgent.notnull()) & (df.InfectiousAgent.str.contains('Culture negative')), 'InfectiousAgent'] = np.NaN
df = df.reset_index(drop=True)

df2 = deepcopy(df)
index_to_remove = []
for k,ia in enumerate(df['InfectiousAgent']):
    if str(ia)=='nan':
        continue
    if '  ' in ia:
        tmps = ia.split('  ')
        for ia2 in tmps:
            df2.loc[len(df2)] = df.iloc[k]
            df2.loc[len(df2)-1,'InfectiousAgent'] = ia2
        index_to_remove.append(k)
df2 = df2.loc[[idx for idx in df2.index if idx not in index_to_remove]]
df2['InfectiousAgent'] = [x.replace('.','').replace('\r','').strip().rstrip('\n').strip().capitalize() if str(x)!='nan' else np.NaN for x in df2['InfectiousAgent']]
df2.to_csv('intermediate_results2.csv', index=False)

df2.head()
```

```python
df = pd.read_csv('intermediate_results3.csv')
df = df[df.InfectiousAgent.notnull()]
df = df.drop(['Full','Positive'], axis=1)

df['AbsDayRelativeToNearestHCT'] = np.abs(df.DayRelativeToNearestHCT)
df = df.sort_values(['PatientID','Timepoint','InfectiousAgent','AbsDayRelativeToNearestHCT'])
df = df.loc[df[['PatientID','Timepoint','InfectiousAgent']].drop_duplicates(keep='first').index].drop('AbsDayRelativeToNearestHCT',axis=1)

df_old = pd.read_csv('../tblInfectionsCidPapers.csv')
df_old.InfectiousAgent = df_old.InfectiousAgent.replace({'Enterococcus_Faecalis':'Enterococcus_faecalis',
                                                         'Enterococcus_Faecium':'Enterococcus_faecium',
                                                         'Enterococcus_Faecium_Vancomycin_Resistant':'Vancomycin_resistant_Enterococcus_faecium',
                                                         'Enterococcus_Vancomycin_Resistant':'Vancomycin_resistant_Enterococcus',
                                                         'Klebsiella_Pneumoniae':'Klebsiella_pneumoniae',
                                                         'Stenotrophomonas_Maltophilia':'Stenotrophomonas_maltophilia'
                                                        })
df_new = pd.merge(df_old, df, left_on=['PatientID','Timepoint','InfectiousAgent','DayRelativeToNearestHCT'], 
                  right_on=['PatientID','Timepoint','InfectiousAgent','DayRelativeToNearestHCT'], how='outer').reset_index(drop=True)
df_new = df_new.drop_duplicates()

index_to_remove = []
for agent in ['Citrobacter','Enterobacter','Escherichia','Klebsiella','Pseudomonas']:
    df_ = df_new[df_new.InfectiousAgent==agent]
    for idx, pid, tps in zip(df_.index, df_.PatientID, df_.Timepoint):
        df__ = df_new[(df_new.PatientID==pid) & (df_new.Timepoint==tps) & (df_new.InfectiousAgent.str.contains(agent))]
        if len(df__)>1:
            index_to_remove.append(idx)
df_new = df_new.loc[[idx for idx in df_new.index if idx not in index_to_remove]]
df_new.InfectiousAgent = [x.replace('_spp','') if x.endswith('_spp') else x for x in df_new.InfectiousAgent]
df_new = df_new.sort_values(['PatientID','Timepoint','InfectiousAgent','DayRelativeToNearestHCT'])

# duplicates found: SD table may have unknown BMT day for some patients but that is known in TR table
df_new = df_new.loc[[x for x in df_new.index if x not in [793, 794, 795, 796, 61, 62]]]
assert len(set(df_new.index)-set(df_new[['PatientID','Timepoint','InfectiousAgent']].drop_duplicates().index))==0

df_new.to_csv('tblInfections.csv', index=False)
```
