```python
import pandas as pd
import numpy as np
from scipy.stats import entropy
from pandas.api.types import is_numeric_dtype
import os
from multiprocessing import Pool
from numpy.random import choice, seed
from datetime import datetime
from scipy.stats import binom
from statsmodels.stats.multitest import fdrcorrection
from functools import partial
import matplotlib.pyplot as plt
from scipy.stats.mstats import gmean
import sys
import seaborn as sns
import warnings
from copy import deepcopy
if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

# Load sample metadata and count table

```python
# read sample meta data and eliminate samples without transplant day
# constrain samples to be collected between [-10, 40]
df_sample = pd.read_csv('tblASVsamples.csv', index_col=0)
df_sample = df_sample[(df_sample.DayRelativeToNearestHCT>=-10) & (df_sample.DayRelativeToNearestHCT<=40)]
print('number of remaining samples = %d'%(len(df_sample)))

# read count data
df_count = pd.read_csv('tblcounts_asv_melt.csv')
df_count = pd.pivot_table(df_count, index='SampleID', columns='ASV', values='Count', aggfunc=np.sum).fillna(0)
sample_ids_w_suff_reads = list(df_count.sum(axis=1)[df_count.sum(axis=1)>=1000].index)
df_count = df_count.loc[set(df_sample.index).intersection(set(sample_ids_w_suff_reads))]
df_relab =  df_count.div(df_count.sum(axis=1), axis=0)

# find commmon samples
common_samples = set(df_sample.index).intersection(set(df_relab.index))
df_sample = df_sample.loc[common_samples]
df_relab = df_relab.loc[common_samples]
```

# Get taxonomy and qPCR, and computer oral fraction

```python
# read taxanomy table
df_tax = pd.read_csv('tblASVtaxonomy_silva138_v4v5_filter.csv', index_col=0)

# read qpcr
df_qpcr = pd.read_csv('tblqpcr.csv', index_col=0)
df_qpcr = df_qpcr[df_qpcr.qPCR16S>0]

# calculate absolute abundance
df_absab = pd.merge(df_relab, df_qpcr, how='inner', left_index=True, right_index=True)
df_absab = df_absab.mul(df_absab['qPCR16S'], axis=0)
df_absab = df_absab.drop(columns=['qPCR16S']).T

# get oral bacterial fraction
df_blast_100 = pd.read_csv("blast_HMPv35oral/blast_HMPv35oral_p100.txt", sep="\t", comment="#", header=None)
df_blast_100.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']#
df_oral_total = df_relab[set(df_blast_100.query_accver).intersection(df_relab.columns)].sum(axis=1).to_frame()
df_oral_total.columns = ['OralFrac_HMPv35oral']
df_oral_total = pd.merge(df_oral_total, df_qpcr, how='inner', left_index=True, right_index=True)
df_oral_total['OralLoad_HMPv35oral'] = df_oral_total['OralFrac_HMPv35oral'].mul(df_oral_total['qPCR16S'], axis=0)
df_oral_total['GutFrac_HMPv35oral'] = 1-df_oral_total['OralFrac_HMPv35oral']
df_oral_total['GutLoad_HMPv35oral'] = df_oral_total['GutFrac_HMPv35oral'].mul(df_oral_total['qPCR16S'], axis=0)
df_oral_total = df_oral_total.drop(columns=['qPCR16S'])
df_oral_total.head()
```

# Run cooccurence analysis

```python
def evaluate(rows, data, sample):
    x = data.loc[rows,:] # extract species in combination
    y = x.min(axis=0)
    y = y[y==1] # cooccur if min of each sample is 1 (that means all species have 1)
    patients = len(np.unique(sample.loc[[z for z in y.index if z in sample.index],'PatientID']))
    observed = len(np.unique(y.index))
    n_samples = data.shape[1]
    probability = (x.sum(axis=1) / n_samples).prod() # probablity if each species enter in community independently
    expected = n_samples * probability
    p_value = 1 - binom.cdf(observed, n_samples, probability) # k, n, p
    return observed, expected, p_value, patients
```

```python
def compute(data,                     # relative abundance data
            sample,                   # sample of patients
            k=100,                    # Compute species sets up to size k
            p=1000,                   # Solution pool size to carry over to next iteration
            abundance_cutoff=0.001,   # Relative abundance cutoff value
            min_samples=1000,         # Cooccurance in at least min_samples samples
            min_patients=100,         # Cooccurance in at least min_patients patients
            fold_cutoff=2,            # Cooccur at least fold_cutoff more than expectede by chance
            fdr_cutoff=0.05,          # FDR-correction threshold
            parallel=True             # Use multi-threaded mode
           ):

    print(f"Inital dataset: {data.shape[0]} species, {data.shape[1]} samples.", flush=True)

    # make sure that data is in relative abundance
    data = data / data.sum(axis=0)
    
    # convert to presence-absence data
    # abundance smaller than threshold is set to 0, otherwise 1
    data = (data >= abundance_cutoff).astype(int)
    
    # data.sum(axis=1) >= min_samples filters for species that occur in at least "min_samples" samples
    # data.sum(axis=0) > 1 filters for samples that have at least two species
    data = data.loc[data.sum(axis=1) >= min_samples, data.sum(axis=0) >= 2]

    print(f"Processed data: {data.shape[0]} species, {data.shape[1]} samples.", flush=True)

    combinations = []
    species = list(data.index.values) # all species names
    pop_sample = [(x,) for x in species] # a tuple

    # "eval_func" would be the same as "evaluate" except that "data" is fixed to current data (remove one variable)
    eval_func = partial(evaluate, data=data, sample=sample)

    if parallel:
        pool = Pool()
        map_func = pool.map
    else:
        map_func = map

    # Compute species sets up to size k
    for i in range(2, k+1):

        print(f"\nComputing co-occurring species sets of size {i}...", flush=True)

        # combine all tuples with one more species (without repetition)
        combinations = (tuple(sorted(set(x)|{y})) for x in pop_sample
                        for y in species if y not in set(x))
        combinations = list(set(combinations))

        # compute scores for all combinations
        scores = map_func(eval_func, combinations)
        scores = pd.DataFrame(scores, columns=["observed", "expected", "p_value", "patients"])
        scores["community"] = combinations
        
        # compute FDR
        scores["significant"], _ = fdrcorrection(scores["p_value"], alpha=fdr_cutoff)

        # filter FDR and min occurrence threshold
        # also control fold change observed/expected >= @fold_cutoff
        selected = scores.query("significant and observed >= @min_samples and patients >= @min_patients and observed >= @fold_cutoff * expected")

        # if no results, stop at this point
        if len(selected) == 0:
            print("No significant hits found.", flush=True)
            break
        else:
            print(f"Selected: {len(selected)} sets.", flush=True)
            selected = selected.sort_values("observed", ascending=False)
            selected = selected[['observed','expected','p_value','patients','community']]
        
        # choose pop_size groups to pass to next iteration
        if p<0:
            pop_size = len(selected)
        else:
            pop_size = min(p, len(selected))
        pop_sample = selected["community"].head(pop_size)
        
        # save communities to splited files
        selected.to_csv('cooccurrence_data/size_%d.csv'%(i))
```

```python
compute(df_relab.T, df_sample, p=-1, min_samples=500, min_patients=200, parallel=False)
```

# Show cooccuring groups and label oral ASV

```python
# read into cooccurrence network
size = [2,3,4,5,6]
df_cooccur_comm = pd.DataFrame()
for s in size:
    df = pd.read_csv('cooccurrence_data/size_%s.csv'%(s), index_col=0).reset_index(drop=True)
    df = df[['community','observed','expected','patients','p_value']]
    df['community'] = [tuple(sorted(eval(x))) for x in df['community']]
    df['fc'] = df['observed']/df['expected']
    df['size'] = s
    df_cooccur_comm = pd.concat([df_cooccur_comm,df]).reset_index(drop=True)

# label whether each network is composed of only oral bacteria
lines = []
df_blast_100 = pd.read_csv("blast_HMPv35oral/blast_HMPv35oral_p100.txt", sep="\t", comment="#", header=None)
df_blast_100.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']#
for asvs in df_cooccur_comm.community:
    if len(set(asvs).intersection(set(df_blast_100.query_accver)))==len(set(asvs)):
        lines.append(True)
    else:
        lines.append(False)
df_cooccur_comm['is_oral_comm'] = lines
print('number of oral communities = %d/%d'%(len(df_cooccur_comm[df_cooccur_comm.is_oral_comm==True]), len(df_cooccur_comm)))

df_cooccur_comm.head()
```
