```python
import pandas as pd
from Bio import Entrez
Entrez.email = 'liaochen1988@gmail.com'
```

# Read strain information

```python
df = pd.read_excel("strain_info.xlsx")
df = df[df['Genbank (16S)'].notnull()]
print("%d sequences found." % (len(df)))
df.head()
```

# Download sequences from NCBI and write to fasta file

```python
out = open('miBC.fasta', 'w')
for acc in df['Genbank (16S)']:
    handle = Entrez.efetch(db="nucleotide", id=acc, rettype="fasta", retmode="text")
    record = handle.read()
    record = record.split('\n')[0] + '\n' + ('').join(record.split('\n')[1:]) + '\n'
    out.write(record)
out.close()
```
